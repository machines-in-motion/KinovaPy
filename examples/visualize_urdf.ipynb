```python
%matplotlib widget
import os
import pinocchio as pin
from pinocchio.visualize import MeshcatVisualizer
from ipywidgets import interact
from KinovaPy import MESHES_PATH, URDF_PATH
```

```python
# Load robot model
urdf_name = 'GEN3-7DOF-VISION_ARM_URDF_V12.urdf'
urdf_path = os.path.join(URDF_PATH, urdf_name)
meshes_dir = MESHES_PATH
robot = pin.RobotWrapper.BuildFromURDF(
            urdf_path, 
            meshes_dir, 
            root_joint=None
        )
rmodel = robot.model
cmodel = robot.collision_model
vmodel = robot.visual_model

# Create Meshcat visualization
viz = MeshcatVisualizer(rmodel, cmodel, vmodel)
try:
    viz.initViewer(open=True)
except ImportError as err:
    print("Error while initializing the viewer. Install Python meshcat?")
    print(err)
    exit(0)
viz.loadViewerModel()

# Initial config
q = pin.neutral(robot.model)
print(f"nq = {len(q)}")
viz.display(q)
```

### Interact
Visualize robot movement by the sliders

```python
def move_joints(j1, j2, j3, j4, j5, j6, j7):
    q = pin.neutral(robot.model)
    q[0] = j1
    q[1] = j2
    q[2] = j3
    q[3] = j4
    q[4] = j5
    q[5] = j6
    q[6] = j7
    viz.display(q)

interact(move_joints, 
         j1=(-1.,1), j2=(-1.,1), j3=(-1.,1), j4=(-1.,1), j5=(-1.,1), j6=(-1.,1), j7=(-1.,1))
```
